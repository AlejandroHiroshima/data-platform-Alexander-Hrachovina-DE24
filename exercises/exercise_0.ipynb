```python
schema = {
    "id": int,
    "name": str,
    "is_active": bool,
    "age": int
}

records = [
    {"id": 101, "name": "Erika", "is_active": True, "age": 45},
    {"id": 102, "name": "Marcus", "is_active": True, "age": 34},
    {"id": 103, "name": "David", "is_active": False, "age": 29},
    {"id": 104, "name": "Anna", "is_active": True, "age": 41.5},
    {"id": 106, "name": "Ingrid", "is_active": "NOPE", "age": 8} 
]

for record in records:
    validation_list = [isinstance(record[key], datatype) for key, datatype in schema.items()]
    is_valid = all(validation_list)

    print(f"Record ID: {record['id']} - Name: {record['name']} - Aktiv: {record['is_active']} - Age: {record['age']}.")
    print(f"Valid: {is_valid}")
```

```python
# Dictionary för att hålla koll på total bytes per protokoll
protocol_bytes = {
    'TCP': 0,
    'UDP': 0
}

# Läs och bearbeta logfilen
with open('data/network.log', 'r') as file:
    for line in file:
        # print(line)
        # Dela upp raden vid '|'
        parts = line.strip().split(' | ')

        # Extrahera relevant information
        # Vi skippar första delen (timestamp) och tar resten
        source = parts[1].split(': ')[1]
        destination = parts[2].split(': ')[1]
        protocol = parts[3].split(': ')[1]
        bytes_size = int(parts[4].split(': ')[1])

        # Skriv ut formaterad rad
        print(f"Source: {source} | Destination: {destination} | Protocol: {protocol} | Bytes: {bytes_size}")

        # Uppdatera totala bytes för protokollet
        protocol_bytes[protocol] += bytes_size

# Skriv ut sammanfattningen
print("\nData Transfer Summary:")
for protocol, bytes_total in protocol_bytes.items():
    print(f"{protocol}: {bytes_total} bytes")
```

```python
import json

with open('data/paid.json', 'r') as file:
    data = json.load(file)
data

```

```python
from collections import defaultdict

pay_sum = defaultdict(int) # Med defaultdict, om nyckeln inte finns, skapas den automatiskt med ett standardvärde (i detta fall 0 eftersom vi använder int).
for i in data: # första iteratione, loopar igenom elementen i listan
    print(i)
    for key, value in i.items(): # andra iterationen kan vi använda .items() eftersom det är dicts, så vi loopar igenom key:value
        print(f"key: {key} value: {value}")
        pay_sum[key] += i[value]
```

**lösningen**


```python
# Iterera över varje dictionary i listan
for record in data:
    # Här är varje "record" en dictionary
    name = record["name"]
    payment = record["paid"]
    pay_sum[name] += payment  # Summera betalningarna för varje namn

# Skriv ut resultatet
for name, total_payment in pay_sum.items():
    print(f"Name: {name}, Total Payment: {total_payment}")

pay_sum = dict(pay_sum) # gör om den till en vanlig dict igen, så den kan bli en json fil

with open ("payment_sum.json", 'w') as json_file:
    json.dump(pay_sum, json_file, indent=4) # json.dump används för att skriva en dict till en fil
```

a) What are the fundamental data types in Python?
The fundamental data types in Python include:

Numeric types: int, float, complex
Sequence types: str, list, tuple, range
Mapping type: dict
Set types: set, frozenset
Boolean type: bool
Binary types: bytes, bytearray, memoryview
None type: NoneType

b) What is the difference between a list and a tuple in Python?
Mutability: Lists are mutable (can be modified), while tuples are immutable (cannot be modified after creation).
Performance: Tuples are generally faster than lists due to their immutability.
Use case: Lists are used for collections of items that may change, while tuples are used for fixed collections of items.

c) How does Python handle mutable and immutable data types?
Mutable types: These can be changed after creation (e.g., list, dict, set). Changes affect the original object.
Immutable types: These cannot be changed after creation (e.g., int, float, str, tuple). Any modification creates a new object.

d) Explain the difference between for and while loops and when to use one or the other.
For loop: Iterates over a sequence (e.g., list, range, string). Use it when the number of iterations is known or when iterating over a collection.
While loop: Repeats as long as a condition is true. Use it when the number of iterations is unknown and depends on a condition.

e) How do you read a file line-by-line in Python?
with open('file.txt', 'r') as file:
    for line in file:
        print(line.strip())

f) What is a context manager and why would you need to use it?
A context manager is a construct in Python that manages resources (e.g., files, database connections) and ensures proper cleanup after use. It is implemented using the with statement. Example:
with open('file.txt', 'r') as file:
    data = file.read()

g) Explain the difference between try, except, else, and finally.
try: Block where you write code that might raise an exception.
except: Block to handle exceptions raised in the try block.
else: Block executed if no exceptions occur in the try block.
finally: Block executed regardless of whether an exception occurred or not.

h) What is the difference between ETL and ELT?
ETL (Extract, Transform, Load): Data is extracted, transformed, and then loaded into the target system.
ELT (Extract, Load, Transform): Data is extracted, loaded into the target system, and then transformed within the system.

i) Differentiate between batch processing and streaming processing.
Batch processing: Processes data in large chunks or batches at scheduled intervals. Suitable for historical data analysis.
Streaming processing: Processes data in real-time as it arrives. Suitable for real-time analytics and monitoring.

j) What is a data platform?
A data platform is an integrated set of tools and technologies that enable the collection, storage, processing, and analysis of data. It supports data pipelines, analytics, and decision-making.

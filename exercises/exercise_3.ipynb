```python
import json

with open('data/orders.json', 'r') as file:
    data = json.load(file)

data
```

```python
for order in data:
    print(f"Input: {order}")
    print(f"Order ID: {order['order_id']}")
    print(f"Kund namn: {order['customer']}")
    order_date = order['order_date']
    order_status = order['status']
    total = 0

    for product in order['products']:
        name = product['name']
        unit_price = product['price']  # Sparar styckpriset
        quantity = product['quantity']

        # Beräkna totalpris för produkten
        total_product_price = unit_price * quantity
        total += total_product_price

        # Visa styckpris och totalpris om quantity > 1
        if quantity > 1:
            print(f"- Produktnamn: {name} (Styckpris: {unit_price:.2f} kr, Totalt: {total_product_price:.2f} kr)")
        else:
            print(f"- Produktnamn: {name} (Pris: {unit_price:.2f} kr)")

        print(f"- Antal: {quantity}")

    print(f"Order date: {order_date}")
    print(f"Order status: {order_status}")
    print(f"Totalsumma: {total:.2f} kr")
    print("---")
 
```
